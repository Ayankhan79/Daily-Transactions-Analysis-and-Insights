# Data Cleaning Process

## 1. Initial Data Inspection

* Loaded the dataset and inspected its structure using `.head()`, `.info()`, and `.describe()`.
* Checked the number of rows and columns.
* Verified the data types of all features.

## 2. Removed Redundant Column

* Removed the **Currency** column because it contained only one unique value (`INR`).
* Since the column had no variability, it did not contribute to the analysis.

## 3. Explored Unique Values

* Examined the unique values of all categorical columns to better understand the dataset.
* This helped identify transaction categories, payment modes, income/expense types, and subcategories.
* It also helped detect inconsistent category names and understand the overall structure of the data.

## 4. Subscription Analysis

* Investigated transactions under the **Subscription** category.
* Identified all subscription services such as Netflix, Amazon Prime, Hotstar, Audible, etc.
* Calculated the total amount spent on each subscription service to understand recurring expenses.

## 5. Date Format Correction

* Discovered that the **Date** column contained mixed formats:

  * `dd/mm/yyyy`
  * `dd/mm/yyyy hh:mm:ss`
* This caused incorrect results when performing date-based operations.
* Split the original column into:

  * **Date** (transaction date)
  * **Time** (transaction timestamp)
* Transactions without timestamps retain `NaT` in the **Time** column.

## 6. Missing Value Treatment

* Analyzed missing values across all columns.
* Replaced missing values in the **Subcategory** column with `"Unknown"` instead of removing the records.
* Retained missing values in the **Time** column because many transactions originally contained only the transaction date.

## 7. Duplicate Record Assessment

* Checked the dataset for duplicate records.
* Identified **9 duplicate rows**.
* Reviewed the duplicate records before deciding whether to remove them, ensuring that genuine repeated transactions were not mistakenly deleted.

## 8. Data Consistency Checks

* Examined categorical values for inconsistencies in naming and capitalization.
* Standardized inconsistent values (for example, `train` → `Train`, `medicine` → `Medicine`, `makeup` → `Makeup`) to improve grouping and reporting accuracy.

## 9. Feature Engineering

Created additional time-based features from the **Date** column:

* **Year**
* **Month Name**
* **Quarter**

These features enable year-wise, monthly, and quarterly trend analysis.

## 10. Final Validation

Performed a final quality check by:

* Verifying data types.
* Reviewing missing values.
* Checking duplicate records.
* Validating unique categorical values.
* Ensuring the dataset was ready for Exploratory Data Analysis (EDA) and Power BI dashboard development.
